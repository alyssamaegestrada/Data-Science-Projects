```python
#Use MathPlotLib

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
#DataSets
#Synthetic Data

np.random.seed(10)

x = np.linspace(1,30,30)
y = np.random.randint(1,60,30)

#Cumulative Data

y = y.cumsum() #Cumulative Sum

iris = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/iris.csv')
```

```python
#1. Draw a Line Graph with figsize (10,6), using x and y variables with custom marker, linestyle and linewidth (3), markersize (8)

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', color = 'black', linestyle='--', linewidth=3, markersize=8)

# Adding labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Graph using MatPlotLib')

# Display the plot
plt.grid(True)
plt.show()
```

```python
#2. Draw a Scatter Graph with figsize (10,6)
plt.figure(figsize=(10, 6))
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], marker='o', label='Scatter Plot', color = 'red')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter Graph Using MatPlotLib')

plt.grid(True)

plt.show()
```

```python
#3
plt.figure(figsize=(10, 6))
x = iris['sepal length (cm)']
y = iris['petal length (cm)']
c = iris['target']

plt.scatter(Iris['sepal length (cm)'], Iris['petal length (cm)'], marker='o', label='Scatter Plot', c = Iris['target'])

# Adding labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter Graph using MatPlotLib')

# Adding a colorbar for target values

# Display the plot
plt.grid(True)
plt.show()
```

```python
#4
#1st Graph
# Plotting the line graph
plt.figure(figsize=(18, 6))

plt.suptitle('Subplotting in MatplotLib')
plt.subplot(1,2,1)
plt.plot(x, y, marker='o', color = 'blue', linestyle='--', linewidth=3, markersize=8)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Graph using MatPlotLib')
plt.grid(True)

plt.legend()

#2nd Graph
plt.subplot(1, 2,2)
iris = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
values = [5.1, 3.5, 1.4, 0.20]
# Create a bar graph with customizations
plt.bar(iris, values, color='red', linewidth=2, alpha=0.9, width=0.8) #barh --> height
plt.title('Subplot - 2')
plt.grid(alpha = 0.5)
plt.ylabel('In Centimeter', fontsize=8)

plt.tight_layout()
plt.legend()
plt.show()
```

```python
#4. Nested Plots

x = np.linspace(1,21,20)
x

np.random.seed(10)

y = np.random.randint(1,50,20)
y

y = y.cumsum()
```

```python
plt.figure(figsize=(12, 5))

# Main graph

plt.plot(x, y,marker = 's', color='green', linestyle='--', linewidth=2)
plt.title('Main Graph')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(False)
plt.legend()

# Secondary Graph
left, bottom, width, height = [0.20, 0.55, 0.30, 0.25]
ax_inset = plt.axes([left, bottom, width, height])
iris = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
values = [5.1, 3.5, 1.4, 0.20]
ax_inset.bar(iris, values, color='red', linewidth=2, alpha=0.9, width=0.8) #barh --> height
ax_inset.set_title('Secondary Graph')
ax_inset.grid(alpha = 0.5)
ax_inset.legend()


#Tertiary Graph
left, bottom, width, height = [0.58, 0.21, 0.30, 0.25]
ax_inset = plt.axes([left, bottom, width, height])
ax_inset.scatter(Iris['sepal length (cm)'], Iris['petal length (cm)'], marker='o', c = Iris['target'])
ax_inset.set_title('Tertiary Graph')
ax_inset.set_xlabel('Sepal Length', fontsize=8)
ax_inset.set_ylabel('Petal Length', fontsize=8)
ax_inset.legend()
# Display the plot
plt.grid(True)
plt.show()
```
