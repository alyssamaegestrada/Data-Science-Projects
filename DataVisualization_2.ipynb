```python
# 5. Nested Plots

# x-axis

x = np.linspace(1,21,20)
x

np.random.seed(10)

y = np.random.randint(1,50,20)
y
y = y.cumsum()
```

```python
# Generate sample data for the main graph

# Create the main graph
plt.figure(figsize=(16, 8))

# Main graph

plt.plot(x, y, color='red', linestyle='-', linewidth=2)
plt.title('Main Graph')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(False)
plt.legend()

# Secondary Graph
left, bottom, width, height = [0.20, 0.63, 0.30, 0.18]
ax_inset = plt.axes([left, bottom, width, height])
ax_inset.plot(x, y, color='green', linestyle='-', linewidth=1)
ax_inset.set_title('Secondary Graph')
ax_inset.set_xlabel('X-axis', fontsize=8)
ax_inset.set_ylabel('Y-axis', fontsize=8)
ax_inset.grid(False)
ax_inset.legend()

#Tertiary Graph
left, bottom, width, height = [0.57, 0.19, 0.30, 0.18]
ax_inset = plt.axes([left, bottom, width, height])
ax_inset.plot(x, y, color='green', linestyle='-', linewidth=1)
ax_inset.set_title('Tertiary Graph')
ax_inset.set_xlabel('X-axis', fontsize=8)
ax_inset.set_ylabel('Y-axis', fontsize=8)
ax_inset.grid(False)
ax_inset.legend()

# Show the plot
plt.show()
```
