```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
#Load DataSet

nba = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/nba%20cleaned.csv')
nba
```

```python
# Connect to an in memory SQLite Database

conn = sqlite3.connect(':memory')

#Write the DataFrame to the Database

nba.to_sql('NBA', conn, index = False)
```

Data Extraction Practice

```python
# 1A. What is the average weight of position of pure Guard (G) for the Miami Heat

query = "SELECT Team, AVG(Weight) FROM NBA WHERE Team = 'Miami Heat' AND Position LIKE '%G%'"
result_sql = pd.read_sql_query(query, conn)
result_sql
```

```python
# 2A. Who is the 1st highest paid player playing dual positions in the NBA

query = "SELECT Name,Position, Salary FROM NBA WHERE Position LIKE '%-%' ORDER BY Salary DESC"
result_sql = pd.read_sql_query(query, conn)
result_sql
```

```python
# 3A. What team has the highest Combine (SUM) Salary in the NBA for players with height of exactly 6-1?
query = "SELECT Team, SUM(Salary) FROM NBA WHERE Height = '6-1'"
result_sql = pd.read_sql_query(query, conn)
result_sql
```

```python
# 4A. Who is the highest paid player that has a Salary of >= 15M, plays at least Center (including dual position players) with Height of 6-10?
query = "SELECT Name, Position, Salary FROM NBA WHERE Salary >= 15000000 AND Position LIKE '%C%' AND Height = '6-10' ORDER BY Salary DESC"
result_sql = pd.read_sql_query(query, conn)
result_sql
```

```python
# 5A. What team has the tallest (Height) and highest paid player in the NBA in ? Note: Zero salaries is not included
query = "SELECT Team, Name, Height, Salary FROM NBA ORDER BY Salary AND Height DESC"
result_sql = pd.read_sql_query(query, conn)
result_sql
```
