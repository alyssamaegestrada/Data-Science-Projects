```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
#Load Dataset

nba = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/nba.csv')
nba
```

```python
#Check for missing data
nba.isnull().sum()
```

```python
#Remove data with all column values of'Nan'

nba = nba.dropna(how = 'all')
nba
```

```python
#Replace Salary with NaN with 0
nba['Salary'].fillna(0, inplace = True)
nba
```

```python
#Replace College with Nan with unknown

nba['College'].fillna('Unknown', inplace = True)
nba
```

```python
#Show actual rows with missing data

nba[nba.isnull().any(axis = 1)]
```

```python
#Remove all data with Nan

nba.dropna(inplace = True)
nba
```

```python
#as type, Salary, Weight data type to int

nba['Salary'] = nba['Salary'].astype('int')
nba['Weight'] = nba['Weight'].astype('int')
nba
```

```python
#Reset Index

nba.reset_index(drop = True, inplace = True)
nba
```

```python
import sqlite3
conn = sqlite3.connect(':memory')
```

```python
nba.to_sql('NBA', conn, index = False)
```

```python

query = "SELECT * FROM NBA"
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#1. What is the average AVG() salary for the Team of 'New York Knicks'

#Execute a SQL Query

query = "SELECT Team, AVG(Salary) AS 'Average Salary of NYK' FROM NBA GROUP BY Team HAVING Team = 'New York Knicks' "
result_sql = pd.read_sql_query(query, conn)
result_sql
```

```python
#2. Who are the top 5 highest paid player on the Miami Heat Team? You can use the keyword LIMIT 5 after the ORDER BY

query = "SELECT Name,Salary FROM NBA WHERE Team = 'Miami Heat'ORDER BY Salary DESC LIMIT 5 "
result_sql = pd.read_sql_query(query, conn)
result_sql
```

```python
#3. Who are the highest paid player for each team in the NBA, sort by Salary Descending? You can use Max(Salary) then GROUP BY

query = "SELECT Name, Team, MAX(Salary) AS 'Highest Paid' FROM NBA GROUP BY Team ORDER BY Salary DESC"
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#4. How many players are part of each Team, ORDER BY 'Current Players' in DESCENDING?

query = "SELECT Team, COUNT(Name) AS 'Current Player' FROM NBA GROUP BY Team ORDER BY COUNT(Name) DESC "
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#5. Who are the players that have salaries between 10M to 25M in the Houston Rockets or Dallas Mavericks, in Salary descending order, Name in ascending order?

query = "SELECT * FROM NBA WHERE (Team = 'Houston Rockets' OR Team = 'Dallas Mavericks') AND Salary BETWEEN 10000000 AND 25000000 ORDER BY Salary DESC, Name ASC;  "
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#6. Show me the highest and lowest paid amount of Salary in the Phoenix Suns roster?

query = "SELECT MAX(Salary) AS 'Highest Salary', MIN(Salary) AS 'Lowest Salary' from NBA WHERE Team = 'Phoenix Suns' "
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#7. The General Manager is looking for any (Forward), it does not matter if he is playing two positions, with weight of 185 lbs to 220 lbs. above, with current salary of 20M to 30M

query = "SELECT * FROM NBA WHERE Position LIKE '%F%' AND Salary BETWEEN 20000000 AND 30000000 AND Weight BETWEEN 185 AND 220"
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#8. Who are the players who entered the league from High School and has a salary of 30M and above, ORDER BY Name

query = "SELECT * FROM NBA WHERE College LIKE '%HS%' AND Salary >=30000000 ORDER BY Name DESC"
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#9. The team is looking for a pure Center (C) with height of at least 7 feet tall, Salary of 18M and above ,  with weight of 250 and above
#Salary in descending ORDER

query = "SELECT * FROM NBA WHERE Position LIKE 'C' AND Height >= 7 AND Salary >= 18000000 AND Weight >= 250 ORDER BY Name DESC "
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#10. Give me players that plays two positions from the Brooklyn Nets of New York Knicks, sort by Salary Descending and Weight Descending

query = "SELECT * FROM NBA WHERE Position LIKE '%-%' AND (Team = 'Brooklyn Nets' OR Team = 'New York Knicks')  ORDER BY Salary DESC, Weight DESC"
result_sql = pd.read_sql_query(query,conn)
result_sql
```

```python
#11. What are the teams in NBA that has an average Salary of 9M and above?

query = "SELECT Team, AVG(Salary) AS 'Average Salary' FROM NBA GROUP BY Team HAVING AVG(Salary)>=9000000 ORDER BY AVG(Salary) DESC"
result_sql = pd.read_sql_query(query,conn)
result_sql
```
