```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore', 'all')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/loan_approval.csv')
df.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
transactions = df.apply(lambda row: row.dropna().tolist(), axis = 1).tolist()
encoder = TransactionEncoder()
transaction_matrix = encoder.fit_transform(transactions)
transaction_df = pd.DataFrame(transaction_matrix, columns = encoder.columns_)
transaction_df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(transaction_df, min_support = 0.1, use_colnames = True)
import warnings
warnings.filterwarnings('ignore', 'all')

frequent_itemsets
```

```python
pd.set_option('display.max_columns', 100)

rules = association_rules(frequent_itemsets, num_itemsets = len(transaction_df), metric = "confidence", min_threshold = 0.1)
rules.loc[:, :'lift']
```

```python
rules.sort_values(by='confidence', ascending=False).head(5)
```

```python
rules.sort_values(by='antecedent support', ascending=False).head(5)
```

```python
rules.sort_values(by='consequent support', ascending=False).head(5)
```

```python
from mlxtend.preprocessing import TransactionEncoder

transactions = df.apply(lambda row: row.dropna().tolist(), axis = 1).tolist()
encoder = TransactionEncoder()
transaction_matrix = encoder.fit_transform(transactions)

transaction_df = pd.DataFrame(transaction_matrix, columns = encoder.columns_)
transaction_df = transaction_df[transaction_df['Loan_Granted'] == True]
transaction_df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(transaction_df, min_support = 0.1, use_colnames = True)
frequent_itemsets
```

```python
if 'support' not in frequent_itemsets.columns:
   frequent_itemsets = frequent_itemsets.rename(columns={'sup': 'support'})
```

```python
pd.set_option('display.max_columns', 10000)

import warnings
warnings.filterwarnings('ignore', 'all')

rules = association_rules(frequent_itemsets, metric ="confidence", min_threshold = 0.1, num_itemsets=100000)
rules.loc[:, :'lift']
```

```python
rules.sort_values(by='confidence', ascending=False).head(5)
```

```python
rules.sort_values(by='antecedent support', ascending=False).head(5)
```

```python
rules.sort_values(by='consequent support', ascending=False).head(5)
```
