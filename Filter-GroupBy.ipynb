```python
import numpy as np
import pandas as pd
import statistics

import warnings
warnings.filterwarnings("ignore")
```

Create a Dataset using this code

```python
#create a sample dataset with 15 rows and 5 columns

np.random.seed(1)

data = {
    'City': ['City_' + str(i) for i in np.random.randint(1,6,15)],
    'Population': [np.random.randint(100000,10000000)for _ in range(15)],
    'Area': [np.random.uniform(1000,10000)for _ in range(15)],
    'GDP': [np.random.uniform(100,500)for _ in range(15)],
    'Capital': ['Capital_' + str(i) for i in np.random.randint(1,5,15)]
}

df = pd.DataFrame(data)
df
```

Using groupby, sort_values, aggregation functions. etc. Do the following:

```python
#Remember the concept of .loc, label based on row and column
#GroupBy 'Capital' and Sum

df.groupby('Capital').sum()
```

```python
#GroupBy 'City' with Mean + specifics

data = df.groupby('City')[['Area','GDP']].mean()
data
```

```python
#GroupBy 'Capital' with Mean, Sort in Ascending Order by Capital + Specifics
data = df.groupby('Capital')[['Population','GDP']].mean().sort_values('Capital', ascending = True)
data
```

```python
#GroupBy 'City' with Mean, Sort in Descending order by GDP + Specifics in terms of row index and column index (remember .loc row and column notation)
data = df.groupby('City')[['Area','GDP']].mean().sort_values('GDP', ascending = False)
data.loc[['City_1','City_4']]
```

```python
#GroupBy 'City' with Median, Sort in Descending Order by Population + Specifics
data = df.groupby("City").median().loc[['City_1','City_3','City_5']]
data['GDP']
```

```python
#Describe the data
df.groupby('Capital').describe()
```

```python
#Describe the data, with Population only
data = df.groupby('Capital')['Population'].describe()
data
```

```python
#Describe the Data, with Area Only
data = df.groupby('Capital')['Area'].describe()
data
```

```python
#Describe the Data with GDP Only
data = df.groupby('Capital')['GDP'].describe().T
data
```

```python
#Describe the data, with GDP only, sort by max descending
data = df.groupby('Capital')['GDP'].describe().sort_values('max', ascending = False)
data
```

```python
#Describe the data, with Area Only, Sort by std descending + Specifics
data = df.groupby(by = "City").describe()['Area'].sort_values(by = "std", ascending = False).loc[['City_1', 'City_3', 'City_5'], ['std', '25%', '75%']]
data
```

```python
#Describe the data, with population only, sort by mean descending + Specifics + Transpose
data = df.groupby(by = "Capital").describe()['Population'].sort_values(by = "mean", ascending = False).loc[['Capital_4', 'Capital_3', 'Capital_1'], ['mean', '50%', 'max']].transpose()
data
```

Handling Missing Data

```python
#Sample Dataset - dictionary into dataframe
d = {'A': [np.nan,20,30,40, np.nan], 'B': [5,5,10,np.nan,15], 'C': [15,np.nan,45,np.nan,30], 'D' :[10,20,40,80,100]}
dataset = pd.DataFrame(d)
dataset
```

```python
#Count the missing data
dataset.isnull().sum()
```

```python
#Drop as long as there are missing data
dataset.dropna()
```

```python
#Drop columns with missing data
dataset.dropna(axis = 1)
```

```python
#Drop with threshold of 4, in terms of columns
dataset.dropna(axis = 1, thresh = 4)
```

```python
#Fill Missing Values of A column using Mean
dataset['A'] = dataset['A'].fillna(value = dataset['A'].mean())
dataset
```

```python
#Fill the Missing Values of C column using Median
dataset['C'] = dataset['C'].fillna(value = dataset['C'].median())
dataset
```

```python
#Fill the Missing Values of B column using Mode
dataset['B'] = dataset['B'].fillna(value = dataset['B'].mode())
dataset
```
