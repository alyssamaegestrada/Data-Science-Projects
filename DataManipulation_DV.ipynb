```python
Dataset = dataset.drop(['Sample code number'], axis = 1)
Dataset
```

```python
Dataset.columns
```

```python
data = [[5, 9, 9, 3, 7, 3, 8, 9, 2, -1],[3, 2, 1, 1, 2, 3, 1, 1, 1, -1]]

validation_data = np.array(data)
validation_data = pd.DataFrame(data, index = [683, 684], columns = Dataset.columns)
validation_data
```

```python
Dataset = pd.concat([Dataset, validation_data], axis = 0)
Dataset.tail(6)
```

```python
from sklearn.preprocessing import StandardScaler

X = Dataset.drop('Class', axis = 1)
X.head(1)
```

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
scaled_features = scaler.transform(X)
scaled_features
```

```python
Dataset.head(1)
```

```python
Dataset.columns[:-1]
```

```python
ds_feat = pd.DataFrame(scaled_features, columns = Dataset.columns[:-1])

ds_feat
```

```python
X = ds_feat.iloc[:683]
y = Dataset.iloc[:683, 9:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state = 101)
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
from sklearn.metrics import classification_report, confusion_matrix

cm = confusion_matrix(y_test, prediction)
sns.heatmap(cm, annot = True, fmt = "d", xticklabels = ['benign', 'Malignant'], yticklabels = ['benign', 'malignant'], cbar = False).set(title = 'Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test,prediction))
```
