```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data

df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/monthly_milk_production.csv')
df.head(10)
```

```python
#Q1
#What is the current data type of the Date column?
#Is it String, datetime64, date, or object?
df.dtypes
```

```python
#Q2
#Using the Date column, change it to date format using .pd.to_datetime() and assign it to df['Date]
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
#Q3
#Set the 'Date' column to be the new  index without using .read_csv, re-assign it to df variable
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
#Plot the DataFrame with a very simple lineplot.
df.plot();
```

```python
#Q5
#Add a column called Month that takes the month (.month) value from the index
df['Month'] = df.index.month
df.head(5)
```

```python
#7 Create a Pandas boxplot by Month with figsize = (12,5)
sns.boxplot(x = 'Month',y = 'Production', figsize = (12,5)
```
